```python
import pandas as pd
import pyodbc

# Connexion à la base de données SQL Server
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=DESKTOP-GQR709G\SQLEXPRESS;DATABASE=PFE_DWH;UID=sa;PWD=213GGMT7258')

# Requête SQL pour récupérer les données
sql_query = "SELECT date_souscription_aff, SUM(mca) as somme_mca FROM dbo.Fact_Affaire GROUP BY date_souscription_aff"

# Charger les données dans un DataFrame
df_CA_Date_sous = pd.read_sql(sql_query, conn)

# Convertir la colonne 'date_souscription_aff' en type datetime64[ns]
df_CA_Date_sous['date_souscription_aff'] = pd.to_datetime(df_CA_Date_sous['date_souscription_aff'])

# Créer un DataFrame avec toutes les dates dans la plage de dates souhaitée
start_date = '2022-10-05'
end_date = '2024-01-06'
date_range = pd.date_range(start=start_date, end=end_date)
df_date_range = pd.DataFrame(date_range, columns=['date_souscription_aff'])

# Fusionner les DataFrames pour remplir les dates manquantes avec NaN
df_CA_Date_souscription = pd.merge(df_date_range, df_CA_Date_sous, on='date_souscription_aff', how='left')

# Exporter le DataFrame vers un fichier Excel
df_CA_Date_souscription.to_excel('C:/Users/dell/OneDrive/Bureau/9raya_esprit/PFE/Fichiers_vue_ensemble/KPIS/df_CA_Date_Sous.xls', index=False)
```
